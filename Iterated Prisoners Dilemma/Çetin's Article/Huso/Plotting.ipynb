## Plotting the Simulation Data

```python
import pandas as pd
import numpy as np
import seaborn as sns; 
from matplotlib import pyplot as plt
```

###Â Experiment Data

```python
df = pd.read_csv("simulation.csv", index_col=None)
```

```python
df.columns = ["(M/N)", "(D/N)","pOfC","pOfD","(pOfC-pOfD)"]
df
```

```python
ddf1 = df[['(M/N)', '(D/N)', 'pOfC']]
ddf1.head()
```

```python
ddf2 = df[['(M/N)', '(D/N)', 'pOfD']]
ddf2.head()
```

```python
ddf3 = df[['(M/N)', '(D/N)', '(pOfC-pOfD)']]
ddf3.head()
```

```python
result = '(pOfC-pOfD)'
plt.figure(figsize=(9,9))
pivot_table = ddf3.pivot('(D/N)','(M/N)',result)
plt.xlabel('D/N', size = 15)
plt.ylabel('M/N', size = 15)
plt.title('pOfC-pOfD', size = 15)
sns.heatmap(pivot_table, annot=True, fmt=".1f", linewidths=.5, square = True, cmap = 'Blues');
```

```python
result = 'pOfC'
plt.figure(figsize=(9,9))
pivot_table = ddf1.pivot('(D/N)','(M/N)',result)
plt.xlabel('D/N', size = 15)
plt.ylabel('M/N', size = 15)
plt.title('pOfC', size = 15)
sns.heatmap(pivot_table, annot=True, fmt=".1f", linewidths=.5, square = True, cmap = 'Blues');
```
